```python
import pandas as pd 
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='Salary', y = 'YearsExperience',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']),y = data['YearsExperience'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['Salary']),y = data['YearsExperience']))
```

```python
data.plot(kind = 'scatter', x='Salary', y = 'YearsExperience',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['Salary']))
plt.plot(data['Salary'],y_pred,color = 'red')
```

```python
## Bac 2 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T 
y = data['YearsExperience']
model.fit(X,y)
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['YearsExperience']))

import numpy as np

range = np.arange(1,200000,1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize=(9,9),color='blue',xlim=(0,200000))
plt.plot(range, y_pred, color = 'red')
```
