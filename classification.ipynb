```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
url = '/content/classification.csv'
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()

correlation = data.corr()
print (correlation)
```

```python
sns.pairplot(data)
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='age', y = 'success',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['age']),y = data['success'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['age']),y = data['success']))
```

```python
data.plot(kind = 'scatter', x='age', y = 'success',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['age']))
plt.plot(data['age'],y_pred,color = 'red')
```

```python
## Bac 2 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'],data['age']**2]).T 
y = data['success']
model.fit(X,y)
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['success']))

import numpy as np

range = np.arange(1,100,0.000001)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'age', y = 'success', figsize=(9,9),color='blue',xlim=(0,100))
plt.plot(range, y_pred, color = 'red')
```

```python
## Bac 3 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'],data['age']**2,data['age']**3]).T
y = data['success'] 
model.fit(X,y )
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['success']))

import numpy as np

range = np.arange(1,100,0.000001)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'age', y = 'success', figsize=(9,9),color='blue',xlim=(0,100))
plt.plot(range, y_pred, color = 'red')
```

```python
## Bac 4
model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'],data['age']**2,data['age']**3,data['age']**4]).T 
model.fit(X,y = data['success'])
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['success']))

import numpy as np

range = np.arange(0,1,0.001)
X_range = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'age', y = 'success', figsize=(9,9),color='blue',xlim=(0,1))
plt.plot(range, y_pred, color = 'red')
```
