```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
url = '/content/housing1.csv'
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()

correlation = data.corr()
print (correlation)
```

```python
sns.pairplot(data)
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='latitude', y = 'households',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['latitude']),y = data['households'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['latitude']),y = data['households']))
```

```python
data.plot(kind = 'scatter', x='latitude', y = 'households',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['latitude']))
plt.plot(data['latitude'],y_pred,color = 'red')
```

```python
## Bac 2 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['latitude'],data['latitude']**2]).T 
y = data['households']
model.fit(X,y)
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['households']))

import numpy as np

range = np.arange(1,50,0.01)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'latitude', y = 'households', figsize=(9,9),color='blue',xlim=(0,100))
plt.plot(range, y_pred, color = 'red')
```

```python
## Bac 3 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['latitude'],data['latitude']**2,data['latitude']**3]).T
y = data['households'] 
model.fit(X,y )
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['households']))

import numpy as np

range = np.arange(0,100,0.01)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'latitude', y = 'households', figsize=(9,9),color='blue',xlim=(0,100))
plt.plot(range, y_pred, color = 'red')
```

```python
## 2 bien

model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['latitude','housing_median_age']],y=data['households'])

print('He so Coeffinet la:',model.coef_)
print('He so Intercept la:',model.intercept_)
print('Do chinh xac cua mo hinh:',model.score(X = data.loc[:,['latitude','housing_median_age']],y=data['households']))

import numpy as np

range = np.arange(0,100,0.01)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)

plt.figure(figsize=(9,9))
plt.scatter(x=data['latitude'], y=data['households'], color='blue', label='latitude')
plt.scatter(x=data['housing_median_age'], y=data['households'], color='green', label='housing_median_age')
plt.plot(range, y_pred, color='red')
plt.xlim((0,100))
plt.xlabel('latitude, housing_median_age')
plt.ylabel('households')
plt.show()
     
```

```python
# 3 bien

model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['latitude','housing_median_age','total_rooms']],y=data['households'])

print('He so Coeffinet la:',model.coef_)
print('He so Intercept la:',model.intercept_)
print('Do chinh xac cua mo hinh:',model.score(X = data.loc[:,['latitude','housing_median_age','total_rooms']],y=data['households']))

import numpy as np

range = np.arange(0,100,0.01)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X_range)

plt.figure(figsize=(9,9))
plt.scatter(x=data['latitude'], y=data['households'], color='blue', label='latitude')
plt.scatter(x=data['housing_median_age'], y=data['households'], color='green', label='housing_median_age')
plt.scatter(x=data['total_rooms'], y=data['households'], color='yellow', label='total_rooms')

plt.plot(range, y_pred, color='red')
plt.xlim((0,100))
plt.xlabel('latitude, housing_median_age,total_rooms')
plt.ylabel('households')
plt.legend()
plt.show()
```
