```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
url = '/content/passenger.csv'
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()

correlation = data.corr()
print (correlation)
```

```python
sns.pairplot(data)
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='Miles(X)', y = 'Dollars (Y)',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Miles(X)']),y = data['Dollars (Y)'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['Miles(X)']),y = data['Dollars (Y)']))
```

```python
data.plot(kind = 'scatter', x='Miles(X)', y = 'Dollars (Y)',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['Miles(X)']))
plt.plot(data['Miles(X)'],y_pred,color = 'red')
     
```

```python
## Bac 2 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2]).T 
y = data['Dollars (Y)']
model.fit(X,y)
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['Dollars (Y)']))

import numpy as np

range = np.arange(1,50000,1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize=(9,9),color='blue',xlim=(0,5000))
plt.plot(range, y_pred, color = 'red')
```
